```python
import math

#get every number that is not NaN
def test(L):
    ret = []
    line = 0
    for j in L:
        column = 0
        for i in j:
            if not math.isnan(i):
                ret.append((line, column, i))
            column += 1
        line += 1 
    return ret


#Format the output (variable 1, variable 2, coefficient of correlation)
def translate(L, columns):
    ret = []
    for i in L:
        if (i[0] != i[1]):
            ret.append((columns[i[0]], columns[i[1]], i[2]))
    return ret

translate(test(tabs), dataset.columns)
```

```python
correlated_parameters = translate(test(tabs), dataset.columns)
```

```python
sorted_by_corr = sorted(correlated_parameters, key=lambda tup: tup[2], reverse=True)
```

```python
def percentage(v1, v2):
    if v1 == v2:
        return 100
    elif v1 == 0:
        return -101
    return v2 * 100 / v1

def changes(data, L1, L2):
    ret = []
    i = 0
    while (i < len(L1)):
        if type(L1[i]) is int and type(L2[i]) is int:
            ret.append((data.columns[i], 100 - percentage(L1[i], L2[i])))
        i+=1
    return ret


changes(day, day.values[0], day.values[1])
```
